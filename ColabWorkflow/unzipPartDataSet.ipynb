```python
import os
import pickle
from typing import List, Tuple
from zipfile import ZipFile
```

```python
path_to_drive = '/content/drive/MyDrive/ImageNetproject/'
```

```python
with open(path_to_drive + 'filtered_data_10class.pkl', 'rb') as file:
    train_data = pickle.load(file)
    val_data = pickle.load(file)
```

```python
path_to_arc = path_to_drive + 'ImageNetDataSet/imagenet-object-localization-challenge.zip'
path_to_data = 'ILSVRC/Data/CLS-LOC/train/'  # path in that archive
```

```python
def unzip_data(data: List[Tuple[str, int]], arc_path: str, data_path: str) -> None:
    """Unzip part of data."""
    with ZipFile(arc_path, 'r') as zip_ref:
        # For the purpose of unzip we do not need class labels
        for file_name, _ in data:
            folder_name = file_name.split('_')[0]
            zip_ref.extract(path_to_data + folder_name + '/' + file_name)
```

```python
unzip_data(train_data, path_to_arc, path_to_data)  # unzip train dataset
```

```python
unzip_data(val_data, path_to_arc, path_to_data)  # unzip validation dataset
```
