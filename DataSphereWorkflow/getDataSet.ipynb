```python
#pragma dataset init ImageNet --size 340Gb

import kaggle
from os import remove
from zipfile import ZipFile

file_name = 'imagenet-object-localization-challenge.zip'
dest_dir = '/home/jupyter/mnt/datasets/ImageNet/'
dest_file = dest_dir + file_name

kaggle.api.competition_download_files('imagenet-object-localization-challenge', path = dest_dir)

with ZipFile(dest_file, 'r') as zip_ref:
    zip_ref.extractall(dest_dir)

remove(dest_file)

# This prevents serialization of temporarily variables
del zip_ref
```
